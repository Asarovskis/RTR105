```python
from numpy import random
```

```python
N = 5000
```

```python
a = 0
b = 5
```

```python
x = random.uniform(a,b,N)
```

```python
k = [0, 0, 0, 0, 0]
for i in range(N):
    if x[i] < 1:
        k[0] = k[0] + 1
    elif x[i] < 2:
        k[1] = k[1] + 1
    elif x[i] < 3:
        k[2] = k[2] + 1
    elif x[i] < 4:
        k[3] = k[3] + 1
    else:
        k[4] = k[4] + 1
print(k)
```

```python
from math import exp
```

```python
y = random.uniform(a,b,N)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funkcija un taas integraalis (laukums starp funkciju un x ass)')
N1 = 0
for i in range(N):
    if y[i] < ((1-x[i])*exp(-x[i])):
        plt.plot(x[i],y[i],'go')
        N1 = N1 + 1
    else:
        plt.plot(x[i],y[i],'ro')
plt.show()
```
